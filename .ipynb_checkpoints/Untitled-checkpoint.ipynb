```python
!pip install tensorflow numpy matplotlib seaborn scikit-learn pandas

import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report, confusion_matrix

# Konfiguracja wyświetlania
plt.style.use('ggplot')

# --- KONFIGURACJA ŚCIEŻKI ---
# Jeśli masz dane lokalnie, wpisz tutaj ścieżkę do folderu głównego (tam gdzie są foldery 'jeans', 'casual_shirts' itp.)
# Przykład: DATA_DIR = "C:/Users/User/Desktop/dataset_clean"
DATA_DIR = "dataset_clean" 

# PARAMETRY GLOBALNE
IMG_SIZE = (224, 224) # Standard dla MobileNetV2/ResNet
BATCH_SIZE = 32
RANDOM_SEED = 42
```

```python
# --- KROK 1: EDA (Exploratory Data Analysis) ---

# 1.1 Wykres liczebności klas
# Pobieramy etykiety z datasetu (może to chwilę potrwać)
labels = []
for _, label in train_ds.unbatch():
    labels.append(label.numpy())

plt.figure(figsize=(10, 6))
sns.countplot(x=labels)
plt.xticks(ticks=range(len(class_names)), labels=class_names, rotation=45)
plt.title("Liczebność klas w zbiorze treningowym")
plt.xlabel("Klasa")
plt.ylabel("Liczba zdjęć")
plt.show()

# 1.2 Wyświetlenie przykładowych zdjęć
plt.figure(figsize=(12, 12))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
plt.show()
```

```python
print("Wczytywanie danych...")

# Zbiór treningowy (80%)
train_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="training",
    seed=RANDOM_SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode='int' # Etykiety jako liczby całkowite (0, 1, 2...)
)

# Zbiór walidacyjny (20%)
val_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="validation",
    seed=RANDOM_SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode='int'
)

class_names = train_ds.class_names
print(f"\nZnalezione klasy ({len(class_names)}): {class_names}")

# Optymalizacja wydajności (cache'owanie danych w pamięci)
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
# 1. Wizualizacja przykładowych zdjęć z każdej klasy
plt.figure(figsize=(12, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
plt.suptitle("Przykłady zdjęć ze zbioru treningowego")
plt.show()

# 2. Rozkład liczebności klas (Sprawdzenie balansu)
# Uwaga: Iteracja po całym zbiorze, aby policzyć elementy (może chwilę potrwać)
labels_list = []
for images, labels in train_ds.unbatch():
    labels_list.append(labels.numpy())

plt.figure(figsize=(10, 6))
sns.countplot(x=labels_list)
plt.xticks(ticks=range(len(class_names)), labels=class_names, rotation=45)
plt.title("Liczebność klas w zbiorze treningowym")
plt.ylabel("Liczba zdjęć")
plt.show()
```

```python
data_augmentation = tf.keras.Sequential([
  layers.RandomFlip("horizontal"),
  layers.RandomRotation(0.1),
  layers.RandomZoom(0.1),
])

# Normalizacja (Rescaling) zostanie dodana jako warstwa wewnątrz modelu
rescale_layer = layers.Rescaling(1./255)
```

```python
def build_custom_cnn():
    model = models.Sequential([
        layers.Input(shape=IMG_SIZE + (3,)),
        data_augmentation, # Augmentacja tylko podczas treningu
        rescale_layer,     # Skalowanie pikseli 0-1
        
        layers.Conv2D(32, 3, padding='same', activation='relu'),
        layers.MaxPooling2D(),
        
        layers.Conv2D(64, 3, padding='same', activation='relu'),
        layers.MaxPooling2D(),
        
        layers.Conv2D(128, 3, padding='same', activation='relu'),
        layers.MaxPooling2D(),
        
        layers.Dropout(0.3), # Regularyzacja
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(len(class_names), activation='softmax') # Wyjście: 8 klas
    ])
    return model

model_cnn = build_custom_cnn()
model_cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model_cnn.summary()

# Trening z Early Stopping
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

print("\n--- Rozpoczynam trening Modelu 1 (Custom CNN) ---")
history_cnn = model_cnn.fit(
    train_ds,
    validation_data=val_ds,
    epochs=20, # Ustaw np. 50, EarlyStopping przerwie wcześniej
    callbacks=[early_stop]
)
```

```python
def build_transfer_model():
    # Pobranie bazy bez górnych warstw klasyfikujących
    base_model = MobileNetV2(input_shape=IMG_SIZE + (3,),
                             include_top=False,
                             weights='imagenet')
    
    base_model.trainable = False # Zamrażamy wagi bazy
    
    model = models.Sequential([
        layers.Input(shape=IMG_SIZE + (3,)),
        data_augmentation,
        # MobileNetV2 oczekuje wejścia [-1, 1], ma wbudowany preprocess, ale użyjmy Rescaling dla pewności lub dedykowanej funkcji
        layers.Rescaling(1./127.5, offset=-1), 
        
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dropout(0.2),
        layers.Dense(len(class_names), activation='softmax')
    ])
    return model

model_tl = build_transfer_model()
model_tl.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("\n--- Rozpoczynam trening Modelu 2 (Transfer Learning) ---")
history_tl = model_tl.fit(
    train_ds,
    validation_data=val_ds,
    epochs=20,
    callbacks=[early_stop]
)
```

```python
def plot_history(history, title):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(len(acc))

    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title(f'{title} - Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title(f'{title} - Loss')
    plt.show()

plot_history(history_cnn, "Model 1: Custom CNN")
plot_history(history_tl, "Model 2: Transfer Learning")
```

```python
# Wybierz lepszy model do analizy (zazwyczaj Transfer Learning)
best_model = model_tl 

print("Generowanie predykcji dla zbioru walidacyjnego...")
y_true = []
y_pred = []

# Iteracja po zbiorze walidacyjnym, aby pobrać etykiety i predykcje
for images, labels in val_ds:
    preds = best_model.predict(images, verbose=0)
    y_true.extend(labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

# Raport klasyfikacji
print("\n--- RAPORT KLASYFIKACJI ---")
print(classification_report(y_true, y_pred, target_names=class_names))

# Macierz pomyłek
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Przewidziana klasa')
plt.ylabel('Prawdziwa klasa')
plt.title('Macierz Pomyłek (Confusion Matrix)')
plt.show()
```

```python
def predict_image(img_path, model):
    img = tf.keras.utils.load_img(
        img_path, target_size=IMG_SIZE
    )
    img_array = tf.keras.utils.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0) # Create a batch

    predictions = model.predict(img_array)
    score = tf.nn.softmax(predictions[0])

    print(
        "To zdjęcie to najprawdopodobniej {} ({:.2f}% pewności)."
        .format(class_names[np.argmax(score)], 100 * np.max(score))
    )
    plt.imshow(img)
    plt.axis("off")
    plt.show()

predict_image("dataset_clean/casual_shirts/img_0006.jpg", best_model)
```
